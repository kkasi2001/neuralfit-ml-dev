```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50, EfficientNetB0
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```python
import zipfile, os

# ✅ Drive에 있는 파일 경로 지정 (문자열로!)
zip_path = "/content/drive/MyDrive/archive.zip"
extract_path = "/content/dataset"

# ✅ 압축 해제
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# ✅ 폴더 구조 확인
for root, dirs, files in os.walk(extract_path):
    level = root.replace(extract_path, '').count(os.sep)
    indent = ' ' * 4 * level
    print(f"{indent}{os.path.basename(root)}/")
    subindent = ' ' * 4 * (level + 1)
    for f in files[:3]:
        print(f"{subindent}{f}")
```

```python
# 데이터셋 경로 설정
train_dir = '/content/dataset/AugmentedAlzheimerDataset'
valid_dir = '/content/dataset/OriginalDataset'

# 이미지 데이터 전처리 및 증강
train_datagen = ImageDataGenerator(
    rescale=1./255,  # 이미지 크기 정규화
    rotation_range=30,  # 회전
    width_shift_range=0.2,  # 수평 이동
    height_shift_range=0.2,  # 수직 이동
    shear_range=0.2,  # 기울기 변환
    zoom_range=0.2,  # 확대/축소
    horizontal_flip=True,  # 수평 뒤집기
    fill_mode='nearest'  # 채우기 방식
)

valid_datagen = ImageDataGenerator(rescale=1./255)

# 데이터셋 로딩
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),  # 이미지 크기 조정
    batch_size=16,
    class_mode='categorical'  # 다중 클래스 분류
)

valid_generator = valid_datagen.flow_from_directory(
    valid_dir,
    target_size=(224, 224),  # 이미지 크기 조정
    batch_size=16,
    class_mode='categorical'
)
```

```python
resnet_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# 모델을 빌드
model_resnet = models.Sequential([
    resnet_model,  # ResNet50 모델을 추가
    layers.GlobalAveragePooling2D(),  # 풀링 레이어
    layers.Dense(4, activation='softmax')  # 출력층 (클래스 수: 4)
])

# 모델 컴파일
model_resnet.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
efficientnet_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# 모델을 빌드
model_efficientnet = models.Sequential([
    efficientnet_model,  # EfficientNetB0 모델을 추가
    layers.GlobalAveragePooling2D(),  # 풀링 레이어
    layers.Dense(4, activation='softmax')  # 출력층 (클래스 수: 4)
])

# 모델 컴파일
model_efficientnet.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# EfficientNetB0 모델을 정의하고 컴파일하는 코드가 이미 실행되었다는 가정 하에 작성됩니다.

# ResNet50 모델 학습
history_resnet = model_resnet.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=10,
    validation_data=valid_generator,
    validation_steps=valid_generator.samples // valid_generator.batch_size,
    verbose=1
)

# EfficientNetB0 모델 학습
# model_efficientnet이 정의되었다고 가정합니다.
history_efficientnet = model_efficientnet.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=10,
    validation_data=valid_generator,
    validation_steps=valid_generator.samples // valid_generator.batch_size,
    verbose=1
)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 # OpenCV

# 함수 정의 예시 (이 코드가 누락된 경우 추가해야 합니다)
def display_gradcam(img_path, heatmap, model_name, alpha=0.4):
    """Grad-CAM 히트맵을 원본 이미지 위에 오버레이하여 표시하는 함수"""
    # 원본 이미지 로드 및 전처리
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 히트맵 크기 조정
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))

    # 히트맵을 컬러맵으로 변환 (예: JET)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    # 이미지와 히트맵을 알파 블렌딩(오버레이)
    superimposed_img = heatmap * alpha + img * (1 - alpha)
    superimposed_img = np.clip(superimposed_img, 0, 255).astype('uint8')

    # 결과 표시
    plt.imshow(superimposed_img)
    plt.title(f"{model_name} Grad-CAM")
    plt.axis('off')
    plt.show()

# 이 함수가 메모리에 정의된 후, Grad-CAM 실행 코드를 다시 실행해 주세요.
```

```python
#CNN 한 번에 정리
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50, EfficientNetB0
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import zipfile
import os

# --- 1. Google Drive 마운트 및 파일 준비 ---
print("--- 1. Google Drive 마운트 및 파일 준비 ---")
try:
    from google.colab import drive
    drive.mount('/content/drive')
except ImportError:
    print("Google Colab 환경이 아닙니다. drive.mount를 건너뜜.")
    # Colab이 아닌 환경에서는 로컬 경로를 사용하도록 설정해야 합니다.

# Drive에 있는 파일 경로 지정 (문자열로!)
# 사용자 설정 필요: 실제 ZIP 파일 경로로 변경하세요.
zip_path = "/content/drive/MyDrive/archive.zip"
extract_path = "/content/dataset"
TARGET_SIZE = (224, 224)
BATCH_SIZE = 16
EPOCHS = 10 # 원본 코드에 10으로 설정되어 있었으나, 학습 시간 고려하여 1로 설정하는 경우가 많습니다.

# 압축 해제
if not os.path.exists(extract_path):
    print(f"'{zip_path}' 파일을 '{extract_path}'로 압축 해제 중...")
    try:
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
        print("압축 해제 완료.")
    except FileNotFoundError:
        print(f"오류: ZIP 파일을 찾을 수 없습니다. 경로를 확인해 주세요: {zip_path}")
        # 프로그램 종료 대신, 이후 단계에서 폴더 경로가 유효한지 확인합니다.
    except Exception as e:
        print(f"오류: 압축 해제 중 오류 발생: {e}")
        # 프로그램 종료 대신, 이후 단계에서 폴더 경로가 유효한지 확인합니다.
else:
    print(f"'{extract_path}' 폴더가 이미 존재하여 압축 해제를 건너뜁니다.")


# 폴더 구조 확인 및 경로 설정
print("\n--- 2. 폴더 구조 확인 및 경로 설정 ---")
# 경로 설정: 압축 해제된 폴더 구조에 맞게 설정해야 합니다.
# (예시: extract_path/AugmentedAlzheimerDataset 와 extract_path/OriginalDataset)
train_dir = os.path.join(extract_path, 'AugmentedAlzheimerDataset')
valid_dir = os.path.join(extract_path, 'OriginalDataset')

print(f"훈련 데이터 경로: {train_dir}")
print(f"검증 데이터 경로: {valid_dir}")

# 간단한 폴더 구조 출력
if os.path.exists(extract_path):
    for root, dirs, files in os.walk(extract_path):
        level = root.replace(extract_path, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f"{indent}{os.path.basename(root)}/")
        subindent = ' ' * 4 * (level + 1)
        for f in files[:1]: # 파일 3개에서 1개로 줄여 출력
            print(f"{subindent}{f}")
        if level >= 2: # 너무 깊은 폴더는 출력하지 않음
            del dirs[:]
    print("-" * 30)
else:
    print("데이터셋 폴더가 존재하지 않아 경로 확인을 건너뜁니다.")


# --- 3. 이미지 데이터 전처리 및 제너레이터 설정 ---
print("\n--- 3. 이미지 데이터 전처리 및 제너레이터 설정 ---")
# 이미지 데이터 전처리 및 증강
train_datagen = ImageDataGenerator(
    rescale=1./255,      # 이미지 크기 정규화
    rotation_range=30,   # 회전
    width_shift_range=0.2, # 수평 이동
    height_shift_range=0.2,# 수직 이동
    shear_range=0.2,     # 기울기 변환
    zoom_range=0.2,      # 확대/축소
    horizontal_flip=True,# 수평 뒤집기
    fill_mode='nearest'  # 채우기 방식
)

valid_datagen = ImageDataGenerator(rescale=1./255)

# 데이터셋 로딩
print("훈련 데이터 로딩 중...")
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=TARGET_SIZE,  # 이미지 크기 조정
    batch_size=BATCH_SIZE,
    class_mode='categorical'  # 다중 클래스 분류
)

print("검증 데이터 로딩 중...")
valid_generator = valid_datagen.flow_from_directory(
    valid_dir,
    target_size=TARGET_SIZE,  # 이미지 크기 조정
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)


# --- 4. ResNet50 모델 정의 및 컴파일 ---
print("\n--- 4. ResNet50 모델 정의 및 컴파일 ---")
resnet_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
#

# 모델 빌드 (Sequential API)
model_resnet = models.Sequential([
    resnet_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(4, activation='softmax') # 클래스 수: 4
])

# 모델 컴파일
model_resnet.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
print("ResNet50 모델 컴파일 완료.")


# --- 5. EfficientNetB0 모델 정의 및 컴파일 ---
print("\n--- 5. EfficientNetB0 모델 정의 및 컴파일 ---")
efficientnet_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
#

# 모델 빌드 (Sequential API)
model_efficientnet = models.Sequential([
    efficientnet_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(4, activation='softmax') # 클래스 수: 4
])

# 모델 컴파일
model_efficientnet.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
print("EfficientNetB0 모델 컴파일 완료.")


# --- 6. 모델 학습 실행 ---
print("\n--- 6. ResNet50 모델 학습 실행 (Epochs: 10) ---")
# 훈련 시간이 오래 걸릴 수 있습니다.
history_resnet = model_resnet.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=10,
    validation_data=valid_generator,
    validation_steps=valid_generator.samples // valid_generator.batch_size,
    verbose=1
)
print("ResNet50 학습 완료.")


print("\n--- 7. EfficientNetB0 모델 학습 실행 (Epochs: 10) ---")
history_efficientnet = model_efficientnet.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=10,
    validation_data=valid_generator,
    validation_steps=valid_generator.samples // valid_generator.batch_size,
    verbose=1
)
print("EfficientNetB0 학습 완료.")


# --- 8. 최종 디렉토리 내용 확인 ---
print("\n--- 8. 최종 디렉토리 내용 확인 ---")
# 이미지 경로가 포함된 디렉토리 확인
image_directory = os.path.join(valid_dir, 'VeryMildDemented')

if os.path.isdir(image_directory):
    print(f"Listing contents of: {image_directory}")
    files = os.listdir(image_directory)
    print(files[:5]) # 파일 5개만 출력
    print(f"(총 {len(files)} 파일)")
else:
    print(f"오류: 경로가 올바르지 않거나 디렉토리를 찾을 수 없습니다: {image_directory}")
```

```python
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras import layers, models, optimizers

# DenseNet121 모델 불러오기 (ImageNet 사전학습 가중치 사용)
base_model_densenet = DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# DenseNet121 모델 빌드
model_densenet = models.Sequential([
    base_model_densenet,  # Pre-trained DenseNet121
    layers.GlobalAveragePooling2D(),  # Global Average Pooling
    layers.BatchNormalization(),  # Batch Normalization
    layers.Dropout(0.5),  # Dropout for regularization
    layers.Dense(4, activation='softmax')  # 출력층 (4-Class 분류)
])

# 모델 컴파일
model_densenet.compile(optimizer=optimizers.Adam(learning_rate=0.0001),
                       loss='categorical_crossentropy',
                       metrics=['accuracy'])
```

```python
from tensorflow.keras.applications import InceptionV3

# InceptionV3 모델 불러오기 (ImageNet 사전학습 가중치 사용)
base_model_inception = InceptionV3(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# InceptionV3 모델 빌드
model_inception = models.Sequential([
    base_model_inception,  # Pre-trained InceptionV3
    layers.GlobalAveragePooling2D(),  # Global Average Pooling
    layers.BatchNormalization(),  # Batch Normalization
    layers.Dropout(0.5),  # Dropout for regularization
    layers.Dense(4, activation='softmax')  # 출력층 (4-Class 분류)
])

# 모델 컴파일
model_inception.compile(optimizer=optimizers.Adam(learning_rate=0.0001),
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])
```

```python
import tensorflow as tf
from tensorflow.keras.applications import DenseNet121, InceptionV3
from tensorflow.keras import layers, models, optimizers
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from tensorflow.keras.preprocessing import image

# ----------------------------------------------------------------------
# 1. 공통 분류기 빌드 함수 정의
# (기존 Sequential 모델의 Global Pooling 이후 부분을 Functional API로 연결하기 위해 사용)
# ----------------------------------------------------------------------
def build_classifier(base_model_output, num_classes=4):
    """모델의 Feature Map 출력(base_model_output)에 분류기를 연결합니다."""
    x = layers.GlobalAveragePooling2D()(base_model_output)  # Global Average Pooling
    x = layers.BatchNormalization()(x)                      # Batch Normalization
    x = layers.Dropout(0.5)(x)                              # Dropout for regularization
    x = layers.Dense(num_classes, activation='softmax')(x)  # 출력층 (4-Class 분류)
    return x

# ----------------------------------------------------------------------
# 2. Grad-CAM 히트맵 생성 함수 (논리 수정)
# ----------------------------------------------------------------------
def make_gradcam_heatmap_fixed(img_array, model, pred_index=None):
    """
    Grad-CAM 논리에 따라 히트맵을 생성합니다.
    모델은 [마지막 Conv Layer 출력, 최종 Classification 예측값] 두 개를 출력해야 합니다.
    """
    with tf.GradientTape() as tape:
        # 모델 출력 받기: (Conv Layer Feature Map, 최종 예측값)
        conv_outputs, predictions = model(img_array)

        # Gradient Tape가 Conv Layer 출력을 추적하도록 합니다.
        tape.watch(conv_outputs)

        # 예측값에서 가장 큰 값의 인덱스 또는 지정된 인덱스를 선택합니다.
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])

        # 해당 클래스의 스코어를 추출합니다.
        class_channel = predictions[0][pred_index]

    # 1. Gradients 계산: Class Score에 대한 Conv Outputs의 Gradient
    # Shape: (1, H, W, C)
    grads = tape.gradient(class_channel, conv_outputs)

    # 2. 채널별 중요도(Weight) 계산: Global Average Pooling
    # Shape: (C,)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # 3. Feature Map에 Weight 적용 (Weighted Sum)
    conv_outputs = conv_outputs[0] # 첫 번째 배치만 가져옵니다. (H, W, C)

    # Feature Map의 각 채널에 해당 채널의 중요도(Weight)를 곱합니다.
    # Shape: (H, W, C)
    weighted_feature_map = conv_outputs * pooled_grads

    # 채널 방향으로 모두 합산하여 2D 히트맵을 생성합니다.
    # Shape: (H, W)
    heatmap = tf.reduce_sum(weighted_feature_map, axis=-1)

    # 4. ReLU 적용 및 정규화
    # ReLU를 적용하여 관심 없는 영역(음수 값)을 제거합니다.
    heatmap = tf.maximum(heatmap, 0)

    # 정규화 (0~1 범위로)
    max_val = tf.reduce_max(heatmap)
    if max_val == 0:
        return heatmap.numpy()

    heatmap /= max_val
    return heatmap.numpy()

# ----------------------------------------------------------------------
# 3. Grad-CAM 시각화 함수
# ----------------------------------------------------------------------
def display_gradcam(img_path, model, last_conv_layer_name, alpha=0.4):
    # 이미지 로드 및 전처리
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    # 모델에 입력할 때 이미지 정규화 (0~1)
    img_array_norm = np.expand_dims(img_array / 255.0, axis=0)

    # Grad-CAM 생성
    # 수정된 함수 사용
    heatmap = make_gradcam_heatmap_fixed(img_array_norm, model)

    # heatmap을 원본 이미지 크기로 업스케일링 및 0~255 사이 이미지로 변환
    heatmap = np.uint8(255 * heatmap)

    # 컬러맵 적용 (JET)
    jet = cm.get_cmap("jet")
    jet_colors = jet(np.arange(256))[:, :3]
    jet_heatmap = jet_colors[heatmap]

    # numpy 배열 -> 이미지로 변환 후, 원본 크기로 리사이즈
    jet_heatmap = image.array_to_img(jet_heatmap)
    jet_heatmap = jet_heatmap.resize((img.size[0], img.size[1]))
    jet_heatmap = image.img_to_array(jet_heatmap)

    # 원본 이미지와 합성
    # superimposed_img = jet_heatmap * alpha + img_array
    # OpenCV의 addWeighted처럼 작동: (1-alpha)*original + alpha*heatmap
    superimposed_img = (1 - alpha) * img_array + alpha * jet_heatmap
    superimposed_img = image.array_to_img(superimposed_img.astype('uint8')) # uint8로 변환

    # 출력
    plt.figure(figsize=(6, 6))
    plt.imshow(superimposed_img)
    plt.axis('off')
    plt.title(f"Grad-CAM on {model.name}")
    plt.show()

# ----------------------------------------------------------------------
# 4. DenseNet121 모델 설정 (Functional API로 Grad-CAM용 재구성)
# ----------------------------------------------------------------------
base_model_densenet = DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# 마지막 Conv Layer를 찾습니다. (Grad-CAM 출력 1)
# DenseNet121의 마지막 Conv/Concat Layer는 'conv5_block16_concat' 또는 'conv5_block16_concat' 다음의 BN/ReLU 레이어입니다.
last_conv_layer_densenet = base_model_densenet.get_layer('conv5_block16_concat')

# 분류기 부분을 base_model 출력에 연결합니다. (Grad-CAM 출력 2)
classifier_output_densenet = build_classifier(base_model_densenet.output)

# Grad-CAM을 위한 Functional API 모델 생성: [Conv Output, Classification Output]
model_densenet_gradcam = models.Model(
    inputs=base_model_densenet.input,
    outputs=[last_conv_layer_densenet.output, classifier_output_densenet],
    name="DenseNet121_GradCAM"
)

# 모델 컴파일 (실제 훈련에 사용될 경우 필요)
model_densenet_gradcam.compile(optimizer=optimizers.Adam(learning_rate=0.0001),
                               loss='categorical_crossentropy',
                               metrics=['accuracy'])

# ----------------------------------------------------------------------
# 5. InceptionV3 모델 설정 (Functional API로 Grad-CAM용 재구성)
# ----------------------------------------------------------------------
base_model_inception = InceptionV3(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# 마지막 Conv Layer를 찾습니다. (Grad-CAM 출력 1)
last_conv_layer_inception = base_model_inception.get_layer('mixed10')

# 분류기 부분을 base_model 출력에 연결합니다. (Grad-CAM 출력 2)
classifier_output_inception = build_classifier(base_model_inception.output)

# Grad-CAM을 위한 Functional API 모델 생성: [Conv Output, Classification Output]
model_inception_gradcam = models.Model(
    inputs=base_model_inception.input,
    outputs=[last_conv_layer_inception.output, classifier_output_inception],
    name="InceptionV3_GradCAM"
)

# 모델 컴파일 (실제 훈련에 사용될 경우 필요)
model_inception_gradcam.compile(optimizer=optimizers.Adam(learning_rate=0.0001),
                                loss='categorical_crossentropy',
                                metrics=['accuracy'])

# ----------------------------------------------------------------------
# 6. Grad-CAM 실행
# ----------------------------------------------------------------------
# 주의: 이 코드는 모델 훈련 없이 ImageNet 가중치를 기반으로 Grad-CAM을 생성합니다.
# 실제 데이터셋에 훈련된 모델 가중치를 로드해야 정확한 분류 결과를 반영할 수 있습니다.

# 예시 이미지 경로 (실제 파일 경로로 변경하세요)
img_path = '/content/dataset/OriginalDataset/VeryMildDemented/verymildDem1553.jpg'
# img_path가 실제 환경에 맞는지 확인해주세요!

try:
    # 1) DenseNet121 Grad-CAM 실행
    print("--- DenseNet121 Grad-CAM 실행 ---")
    display_gradcam(
        img_path=img_path,
        model=model_densenet_gradcam,
        last_conv_layer_name='conv5_block16_concat',
        alpha=0.4
    )

    # 2) InceptionV3 Grad-CAM 실행
    print("--- InceptionV3 Grad-CAM 실행 ---")
    display_gradcam(
        img_path=img_path,
        model=model_inception_gradcam,
        last_conv_layer_name='mixed10',
        alpha=0.4
    )
except FileNotFoundError:
    print(f"\n오류: 이미지 파일 경로를 찾을 수 없습니다. 경로: {img_path}")
    print("사용하시는 환경에 맞게 img_path 변수를 수정해 주세요.")
except Exception as e:
    print(f"\nGrad-CAM 실행 중 예상치 못한 오류가 발생했습니다: {e}")
```

```python
import tensorflow as tf
import numpy as np

def preprocess_image(img_path, target_size=(224, 224)):
    """이미지 경로를 받아 모델 입력 형태의 텐서로 변환하고 정규화하는 함수"""

    # 1. 이미지 로드 및 크기 조정
    # tf.keras.utils.load_img 또는 tf.io.read_file 등으로 이미지를 로드합니다.
    img = tf.keras.utils.load_img(img_path, target_size=target_size)

    # 2. 이미지를 NumPy 배열로 변환
    img_array = tf.keras.utils.img_to_array(img)

    # 3. 배치 차원 추가 (모델은 [배치 크기, 높이, 너비, 채널] 형태를 요구)
    img_array = np.expand_dims(img_array, axis=0)

    # 4. 정규화 (0~1 사이 값으로 변환)
    return img_array / 255.0

# 이 함수를 정의한 후, Grad-CAM 시각화 코드를 다시 실행해야 합니다.
```

```python
# ResNet50 모델을 Grad-CAM에 맞게 변환하는 예시

# 1. 마지막 컨볼루션 레이어 이름 지정
LAST_CONV_LAYER_NAME = 'conv5_block3_out'
CLASSIFICATION_LAYER_NAME = 'predictions' # 최종 출력 레이어 이름

# 2. Grad-CAM에 필요한 3가지 모델 생성
# (a) 최종 모델: 원본 ResNet50 모델
grad_model = resnet50_model

# (b) 클래스 예측 모델: 최종 출력 레이어만 따로 분리 (클래스 확률 계산용)
classifier_layer = grad_model.get_layer(CLASSIFICATION_LAYER_NAME)
classifier_model = Model(grad_model.inputs, classifier_layer.output)

# (c) 특징 맵 추출 모델: 마지막 컨볼루션 레이어 출력까지 분리
last_conv_layer = grad_model.get_layer(LAST_CONV_LAYER_NAME)
feature_map_model = Model(grad_model.inputs, last_conv_layer.output)

# 이제 이 3가지 모델(grad_model, classifier_model, feature_map_model)을
# 사용하여 이미지와 최종 출력(클래스 점수), 그리고 특징 맵(Feature Map)을 얻어
# Grad-CAM 히트맵을 계산할 수 있습니다.
```
